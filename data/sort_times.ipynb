list_foldernamesをソートして，list_timesというテキストファイルに出力するためのpythonスクリプト

```python
import pandas as pd
```

```python
df_times = pd.DataFrame()

with open('list_foldernames') as f:
   for line in f:
      print(line)

      time_str = line.split('\n')[0]
      time_float = float(time_str)

      ditc_time = {time_str: time_float}
      df_time = pd.DataFrame(ditc_time, index=[0])

      df_times = pd.concat([df_times, df_time], axis=1)

df_times
```

```python
list_times = df_times.sort_values(by=0, axis=1).columns.to_list()
len(list_times)
```

```python
with open('list_times', "w") as f:

    for time in list_times:

        f.write(time+'\n')
```
